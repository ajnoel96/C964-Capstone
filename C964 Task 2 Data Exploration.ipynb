```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


data = pd.read_csv('insurance_claims.csv')

data.head()
```

```python
data.describe()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='age', y='total_claim_amount', data=data)
plt.title('Total Claim Amount vs. Age')
plt.xlabel('Age')
plt.ylabel('Total Claim Amount')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
avg_claim_by_incident = data.groupby('incident_type')['total_claim_amount'].mean().reset_index()
sns.barplot(x='incident_type', y='total_claim_amount', data=avg_claim_by_incident)
plt.title('Average Claim Amount by Incident Type')
plt.xlabel('Incident Type')
plt.ylabel('Average Total Claim Amount')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(12, 8))
sns.boxplot(x='policy_state', y='total_claim_amount', data=data)
plt.title('Total Claim Amount by Policy State')
plt.xlabel('Policy State')
plt.ylabel('Total Claim Amount')
plt.show()
```
